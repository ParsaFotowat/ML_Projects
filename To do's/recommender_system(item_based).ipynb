```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib
```

```python
colums_names=["user_id","item_id","ratings","timestamps"]
```

```python
df=pd.read_csv("u.data.csv",sep="\t",names=colums_names)
```

```python
df.head()
```

```python
movie_id_title=pd.read_csv("Movie_Id_Ti.csv")
```

```python
movie_id_title.head
```

```python
df=pd.merge(df,movie_id_title,on="item_id")
```

```python
df.head
```

```python
sns.set_style("white")
```

```python
df.groupby("title")["ratings"].mean().sort_values(ascending=False).head()
```

```python
df.groupby("title")["ratings"].count().sort_values(ascending=False).head()
```

```python
ratings=pd.DataFrame(df.groupby("title")["ratings"].mean())
```

```python
ratings["# of ratings"]=pd.DataFrame(df.groupby("title")["ratings"].count())
```

```python
ratings.head
```

```python
sns.jointplot(x="ratings",y="# of ratings",data=ratings,alpha=0.5)
```

```python
moviemat = df.pivot_table(index='user_id',columns='title',values='ratings')
moviemat.head()
```

```python
ratings.sort_values('# of ratings',ascending=False).head(10)
```

```python
starwars_user_ratings = moviemat['Star Wars (1977)']
liarliar_user_ratings = moviemat['Liar Liar (1997)']
starwars_user_ratings.head()
```

```python
similar_to_starwars = moviemat.corrwith(starwars_user_ratings)
similar_to_liarliar = moviemat.corrwith(liarliar_user_ratings)
```

```python
corr_starwars = pd.DataFrame(similar_to_starwars,columns=['Correlation'])
corr_starwars.dropna(inplace=True)
corr_starwars.head()
```

```python
corr_starwars.sort_values('Correlation',ascending=False).head(10)
```

```python
corr_starwars = corr_starwars.join(ratings['# of ratings'])
corr_starwars.head()
```

```python
corr_starwars[corr_starwars['# of ratings']>100].sort_values('Correlation',ascending=False).head()
```

```python
corr_liarliar=pd.DataFrame(similar_to_liarliar,columns=["Correlation"])
```

```python
corr_liarliar.dropna(inplace=True)
```

```python
corr_liarliar = corr_liarliar.join(ratings['# of ratings'])
corr_liarliar[corr_liarliar['# of ratings']>100].sort_values('Correlation',ascending=False).head()
```
